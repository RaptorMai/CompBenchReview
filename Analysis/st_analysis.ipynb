# sanity check

```python
refine = []
rerun = []
for pair in gpt:
    new = deepcopy(pair)
    if pair['gpt_answer'] == None :
        rerun.append(pair)
    elif 'Right' in pair['gpt_answer']:
        new['gpt_answer'] = 'Right'
        refine.append(new)
    elif 'Left' in pair['gpt_answer']:
        new['gpt_answer'] = 'Left'
        refine.append(new)
    else:
        refine.append(pair)
        print(pair)
        
```

```python
type_list = []
for pair in gpt:
    type_list.append(pair['type'])
```

```python
Counter(type_list)
```

```python
attribute = ['Size', 'Color', 'Pattern', 'Texture', 'Shape']
```

```python
state_error = []
att_error = []
pred = []
gt = []
state_correct = []
state_wrong = []

att_correct = []
att_wong = []

for pair in gpt:
    if pair['gpt_answer'] not in ('Left', "Right"):
        if pair['type'] in attribute:
            att_error.append(pair)
        elif pair['type'] == 'State':
            state_error.append(pair)
        continue
    if pair['answer'] == pair['gpt_answer']:
        if pair['type'] in attribute:
            att_correct.append(pair)
        elif pair['type'] == 'State':
            state_correct.append(pair)
    elif pair['answer'] != pair['gpt_answer']:
        if pair['type'] in attribute:
            att_wong.append(pair)
        elif pair['type'] == 'State':
            state_wrong.append(pair)
    else:
        print('fuck')
```

```python
len(state_correct) / (len(state_correct) + len(state_wrong))
```

```python
len(att_correct) / (len(att_correct) + len(att_wong))
```

```python
sizes = [len(state_correct), len(state_wrong), len(state_error)]
# Labels with percentages and exact counts
labels = [f'Correct - {sizes[0]} items ({sizes[0]/sum(sizes)*100:.2f}%)',
          f'Wrong - {sizes[1]} items ({sizes[1]/sum(sizes)*100:.2f}%)', 
          f'Error - {sizes[2]} items ({sizes[2]/sum(sizes)*100:.2f}%)', 
         ]

# Colors for each section
colors = ['gold', 'yellowgreen', 'lightcoral']

# Plotting the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('ST State GPT')
plt.show()
```
