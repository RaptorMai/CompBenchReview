```python
import os
import json
from collections import defaultdict
from collections import Counter
from copy import deepcopy
import matplotlib.pyplot as plt
```

# GPT

```python
error = []
refine = []
rerun = []
for pair in gpt:
    new = deepcopy(pair)
    if pair['gpt_answer'] == None :
        rerun.append(pair)
    elif 'Right' in pair['gpt_answer']:
        new['gpt_answer'] = 'Right'
        refine.append(new)
    elif 'Left' in pair['gpt_answer']:
        new['gpt_answer'] = 'Left'
        refine.append(new)
    else:
        error.append(pair)
        print(pair)
        
```

```python
len(refine)
```

```python
len(error)
```

```python
correct_pred = []
wrong_pred = []
gt = []
correct = []
wrong = []

for pair in refine:
    gt.append(pair['answer'])
    # if pair['gpt_answer'] not in ('Left', "Right"):
    #     # print(pair)
    #     error.append(pair)
    #     continue
    if pair['answer'] == pair['gpt_answer']:
            correct.append(pair)
            correct_pred.append(pair['gpt_answer'])
    elif pair['answer'] != pair['gpt_answer']:
            wrong.append(pair)
            wrong_pred.append(pair['gpt_answer'])
    else:
        print('fuck')
```

```python
len(correct) / (len(correct) + len(wrong))
```

```python
Counter(wrong_pred+correct_pred)
```

```python
len(error)
```

```python
(len(correct) + len(wrong))
```

```python
sizes = [len(correct), len(wrong), len(error)]
# Labels with percentages and exact counts
labels = [f'Correct - {sizes[0]} items ({sizes[0]/sum(sizes)*100:.2f}%)',
          f'Wrong - {sizes[1]} items ({sizes[1]/sum(sizes)*100:.2f}%)', 
         f'Error - {sizes[2]} items ({sizes[2]/sum(sizes)*100:.2f}%)', 
         ]

# Colors for each section
colors = ['gold', 'yellowgreen', 'lightcoral']

# Plotting the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('VAW GPT')
plt.show()
```

```python
len(gpt)
```

```python
# Attribute result
att_error = []
att_correct_pred = []
att_wrong_pred = []
att_gt = []
att_correct = []
att_wrong = []

att_lst = ["Size", "Color", "Texture", "Shape", "Pattern"]

#for pair in refine:
# include error
for pair in refine:
    if pair["type"] in att_lst:
        att_gt.append(pair['answer'])
        if pair['gpt_answer'] not in ('Left', "Right"):
            # print(pair)
            att_error.append(pair)
            continue
        if pair['answer'] == pair['gpt_answer']:
                att_correct.append(pair)
                att_correct_pred.append(pair['gpt_answer'])
        elif pair['answer'] != pair['gpt_answer']:
                att_wrong.append(pair)
                att_wrong_pred.append(pair['gpt_answer'])
        else:
            print('fuck')

for pair in error:
    if pair["type"] in att_lst:
        att_error.append(pair)
```

```python
sizes = [len(att_correct), len(att_wrong), len(att_error)]
# Labels with percentages and exact counts
labels = [f'Correct - {sizes[0]} items ({sizes[0]/sum(sizes)*100:.2f}%)',
          f'Wrong - {sizes[1]} items ({sizes[1]/sum(sizes)*100:.2f}%)', 
         f'Error - {sizes[2]} items ({sizes[2]/sum(sizes)*100:.2f}%)', 
         ]

# Colors for each section
colors = ['gold', 'yellowgreen', 'lightcoral']

# Plotting the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('VAW Attribute GPT')
plt.show()
```

```python
# State result
state_error = []
state_correct_pred = []
state_wrong_pred = []
state_gt = []
state_correct = []
state_wrong = []

state_lst = ["State"]

#for pair in refine:
# include error
for pair in gpt:
    if pair["type"] in state_lst:
        state_gt.append(pair['answer'])
        if pair['gpt_answer'] not in ('Left', "Right"):
            # print(pair)
            state_error.append(pair)
            continue
        if pair['answer'] == pair['gpt_answer']:
                state_correct.append(pair)
                state_correct_pred.append(pair['gpt_answer'])
        elif pair['answer'] != pair['gpt_answer']:
                state_wrong.append(pair)
                state_wrong_pred.append(pair['gpt_answer'])
        else:
            print('fuck')

for pair in error:
    if pair["type"] in state_lst:
          state_error.append(pair)
```

```python
len(state_correct) + len(state_wrong) + len(state_error)
```

```python
sizes = [len(state_correct), len(state_wrong), len(state_error)]
# Labels with percentages and exact counts
labels = [f'Correct - {sizes[0]} items ({sizes[0]/sum(sizes)*100:.2f}%)',
          f'Wrong - {sizes[1]} items ({sizes[1]/sum(sizes)*100:.2f}%)', 
         f'Error - {sizes[2]} items ({sizes[2]/sum(sizes)*100:.2f}%)', 
         ]

# Colors for each section
colors = ['gold', 'yellowgreen', 'lightcoral']

# Plotting the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('VAW State GPT')
plt.show()
```
